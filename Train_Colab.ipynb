# Currency Detection YOLO Training (Google Drive Edition)

Since the unified dataset contains over 32,000 files, direct uploading through the Colab side-panel may crash your browser. Using Google Drive is the safest and fastest way!

**Before you start:**
1. Upload the finished `fast_dataset_upload.zip` directly into your Google Drive (in the main root folder).
2. In the Colab menu, go to **Runtime** > **Change runtime type**, and select **T4 GPU**.

```python
# 1. Install YOLO and dependencies
import os
os.system('pip install ultralytics')
```

```python
# 2. Mount Google Drive to access the Zip file
from google.colab import drive
drive.mount('/content/drive')
```

```python
# 3. Extract the Mega-Dataset directly from your Drive
import os
print("Extracting dataset... This may take a quick minute.")
os.system('unzip -q /content/drive/MyDrive/fast_dataset_upload.zip -d /content/')
print("Done!")
```

```python
# 4. Fix absolute paths in data.yaml for Google Colab
import yaml
with open('/content/dataset/data.yaml', 'r') as f:
    data = yaml.safe_load(f)

data['path'] = '/content/dataset'  # Point to Colab's local disk

with open('/content/dataset/data.yaml', 'w') as f:
    yaml.dump(data, f)

print("data.yaml correctly updated!")
```

```python
# 5. Start the Heavy-Duty Training!
from ultralytics import YOLO

# Build from YOLO11m (Medium model)
model = YOLO('yolo11m.pt')

# Train
results = model.train(data='/content/dataset/data.yaml', epochs=120, imgsz=960, batch=8, device=0)
```
